# **LIST WITH CONDITIONS**

List = [**expression** for **item** in **interable** if** condition**]




* expression = x , x*x , x*x*x
* item = variable(for loop)
* interable = range or other things
* coonditions = any type of  conditions like {if x%2 == o}



```python
list1 = [x for x in range(10)]
```

```python
list1
```




---



```python
list2 = [x*x for x in range(10)]
```

```python
list2
```



---
