```python
x = range(0,10)
```

```python
for n in x:    #1st case
  print(n)
```

```python
x = range(5,20)  #2nd case

for n in x:
  print(n)
```

```python
x = range(5,20,3)  #3rd case

for n in x:
  print(n)
```

```python
x = range(20,5,-3)  #4th case

for n in x:
  print(n)
```
